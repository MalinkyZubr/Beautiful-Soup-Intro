# Navigating the HTML Tree

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = r'https://coinmarketcap.com/'
result = requests.get(url).text
doc = BeautifulSoup(result, "html.parser")
```

```python
tbody = doc.tbody
print(tbody) # get the crypto table
```

```python
# I could just get the contents and loop through the table
# we could also use tags to figure out the siblings of the tree
# from any of the table rows I can go to the next or previous sibling
# the tags dont need to be named
# we just go through branches
trs = tbody.contents # make a table row list
print(trs)
```

```python
# now lets get the next sibling from the 0th element in the table
print(trs[0].next_sibling)
# we know its correct because ethereum comes after bitcoin
```

```python
print(trs[1].previous_sibling)
```

```python
# print the parent
print(trs[0].parent)
```

```python
# print the descendants
print(list(trs[0].descendants)) # returns all level under
```

```python
# print the children
print(list(trs[0].children)) # only returns one level under
```

```python
# find all coin names and prices
prices = {}
for tr in trs[:10]: # first ten currencies 
    name, price = tr.contents[2:4] # name and price of each row
    name = name.p.string
    price = price.a.string
     
    prices[name] = price

print(prices)
```
