# Multiple websites, finding the best GPU prices
we will becomparing graphics cards on newegg

what we want to do. Output a list that contains for each graphics card:
1. the name of the gpu
2. the price
3. the link

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
product = str(input("What product are you searching for?: "))
url = f'https://www.newegg.ca/p/pl?d={product}&N=4131' # in the url the d parameter specifies a search on newegg

page = requests.get(url).text

doc = BeautifulSoup(page, "html.parser")
```

```python
# first we need to find how many pages we have, because we want to look on all the pages
# we want to find on the bottom of the page how many pages we have, and loop through all the pages
# using inspect element is useful

page_text = str(doc.find(class_='list-tool-pagination-text').strong)
pages = [page for page in page_text if page.isnumeric()][1]
print(pages)
print(page_text)
```

```python
for page in range(1, int(pages) + 1): # because the pages go 1 to pages -1
    items_found = {}
    url = f'https://www.newegg.ca/p/pl?d={product}&N=4131&page={page}' # in the url the d parameter specifies a search on newegg
    page = requests.get(url).text
    doc = BeautifulSoup(page, "html.parser")
    div = doc.find(class_="item-cells-wrap border-cells items-grid-view four-cells expulsion-one-cell")
    items = div.find_all(text=re.compile(product)) # we want to match any text that contains this. Otherwise, it would only match only the results with text that equals 3080
    for item in items:
        parent = item.parent
        link = None
        if parent.name != 'a':
            continue
        
        next_parent = item.find_parent(class_="item-container")
        price = next_parent.find(class_="price-current").strong.string
        link = parent['href']
        print(f'{item}: ${price} AT LINK {link}')
        items_found[item] = {"Price": int(price), "link": link}
```
